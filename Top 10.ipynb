```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_movie_dataset.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
df.columns
```

```python
top_10_movies = (
    df.groupby("title")["vote_average"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
top_10_movies
```

```python
top_10_movies = top_10_movies.reset_index()
top_10_movies
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=top_10_movies,
    x="vote_average",
    y="title"
)

plt.title("Top 10 Movies by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Movie Title")
plt.show()
```

The bar chart displays the top 10 movies based on their average user ratings.
It shows that these movies have higher ratings compared to others in the dataset.
This visualization helps in identifying the highest-rated movies and comparing
their performance easily.
